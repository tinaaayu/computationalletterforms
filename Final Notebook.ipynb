```python
import tkinter as tk
import nltk
import random

# main window
root = tk.Tk()
root.title("cheese poem")
root.geometry("600x400")  # window size

label = tk.Label(root, text="enter your text below:")
label.pack()

# text entry
entry = tk.Entry(root, width=60)
entry.pack(pady=10)

def generate_text():
    user_input = entry.get()
    if not user_input.strip():
        text_area.delete("1.0", tk.END)
        text_area.insert(tk.END, "enter some text!")
        return

    TEXT = user_input.split()

    if len(TEXT) < 20:
        text_area.delete("1.0", tk.END)
        text_area.insert(tk.END, "please enter at least 20 words.")
        return

    trigrams = list(nltk.trigrams(TEXT))

    cfd = nltk.ConditionalFreqDist(
        (w1_w2, w3) for w1_w2, w3 in (((w1, w2), w3) for w1, w2, w3 in trigrams)
    )

    starting_pair = random.choice(list(cfd.conditions()))
    w1, w2 = starting_pair

    output = [w1, w2]

    for _ in range(50):
        if (w1, w2) in cfd:
            w3 = random.choice(list(cfd[(w1, w2)].keys()))
            output.append(w3)
            w1, w2 = w2, w3
        else:
            w1, w2 = random.choice(list(cfd.conditions()))
            output.extend([w1, w2])

    generated_text = ' '.join(output)

    text_area.delete("1.0", tk.END)  # clear previous output
    text_area.insert(tk.END, generated_text)

button = tk.Button(root, text="generate found poem", command=generate_text)
button.pack(pady=0)

text_area = tk.Text(root, wrap=tk.WORD, height=15, width=75)
text_area.pack(pady=20)

root.mainloop()
```

```python
import tkinter as tk
from tkinter import filedialog, messagebox
import nltk
import random
import threading
import time

# main window
root = tk.Tk()
root.title("🧀 Cheese Poem Generator 🧀")
root.geometry("700x550")
root.configure(bg="cornsilk")
root.resizable(False, False)

top_frame = tk.Frame(root, bg="cornsilk")
top_frame.pack(pady=10)

middle_frame = tk.Frame(root, bg="cornsilk")
middle_frame.pack(pady=10)

bottom_frame = tk.Frame(root, bg="cornsilk")
bottom_frame.pack(pady=10)

# label
label = tk.Label(top_frame, text="Enter your text below:", font=("Helvetica", 16, "bold"), bg="cornsilk")
label.pack()

# text entry
entry = tk.Entry(top_frame, width=70, font=("Helvetica", 12))
entry.pack(pady=8)

# word count
word_count_label = tk.Label(top_frame, text="Words: 0", font=("Helvetica", 10), bg="cornsilk")
word_count_label.pack()

# output length selector
length_label = tk.Label(top_frame, text="Poem Length (words):", font=("Helvetica", 10), bg="cornsilk")
length_label.pack(pady=(10, 0))
length_var = tk.IntVar(value=50)
length_entry = tk.Entry(top_frame, textvariable=length_var, width=10, font=("Helvetica", 12))
length_entry.pack()

# text area
text_area = tk.Text(middle_frame, wrap=tk.WORD, height=15, width=80, font=("Helvetica", 12), bg="white", fg="black")
text_area.pack()


def update_word_count(event=None):
    user_input = entry.get()
    word_count = len(user_input.split())
    word_count_label.config(text=f"Words: {word_count}")

def type_text(text):
    text_area.delete("1.0", tk.END)
    for char in text:
        text_area.insert(tk.END, char)
        text_area.update()
        time.sleep(0.02)  # typing speed

def generate_text():
    user_input = entry.get()
    if not user_input.strip():
        messagebox.showerror("Input Error", "Please enter some text!")
        return

    TEXT = user_input.split()

    if len(TEXT) < 20:
        messagebox.showerror("Input Error", "Please enter at least 20 words.")
        return

    try:
        output_length = int(length_var.get())
    except ValueError:
        messagebox.showerror("Input Error", "Poem length must be a number.")
        return

    trigrams = list(nltk.trigrams(TEXT))

    cfd = nltk.ConditionalFreqDist(
        (w1_w2, w3) for w1_w2, w3 in (((w1, w2), w3) for w1, w2, w3 in trigrams)
    )

    starting_pair = random.choice(list(cfd.conditions()))
    w1, w2 = starting_pair

    output = [w1, w2]

    for _ in range(output_length - 2):
        if (w1, w2) in cfd:
            w3 = random.choice(list(cfd[(w1, w2)].keys()))
            output.append(w3)
            w1, w2 = w2, w3
        else:
            w1, w2 = random.choice(list(cfd.conditions()))
            output.extend([w1, w2])

    generated_text = ' '.join(output)

    # --- Format like a real poem ---
    sentences = []
    current_sentence = ""

    for word in generated_text.split():
        current_sentence += word + " "
        if word.endswith(('.', '!', '?')):
            sentences.append(current_sentence.strip())
            current_sentence = ""

    # If there's leftover text not ending with punctuation
    if current_sentence.strip():
        sentences.append(current_sentence.strip())

    final_output = '\n'.join(sentences)

    # typing animation
    threading.Thread(target=type_text, args=(final_output,), daemon=True).start()

def clear_all():
    entry.delete(0, tk.END)
    text_area.delete("1.0", tk.END)
    word_count_label.config(text="Words: 0")

def save_text():
    file_path = filedialog.asksaveasfilename(defaultextension=".txt", 
                                             filetypes=[("Text files", "*.txt")])
    if file_path:
        with open(file_path, 'w', encoding='utf-8') as file:
            file.write(text_area.get("1.0", tk.END))

# buttons
button_frame = tk.Frame(bottom_frame, bg="cornsilk")
button_frame.pack()

generate_button = tk.Button(button_frame, text="Generate Poem", command=generate_text, 
                            font=("Helvetica", 12, "bold"), bg="lightblue", width=15)
generate_button.grid(row=0, column=0, padx=10, pady=10)

clear_button = tk.Button(button_frame, text="Clear All", command=clear_all,
                         font=("Helvetica", 12, "bold"), bg="lightpink", width=15)
clear_button.grid(row=0, column=1, padx=10, pady=10)

save_button = tk.Button(button_frame, text="Save Poem", command=save_text,
                        font=("Helvetica", 12, "bold"), bg="lightgreen", width=15)
save_button.grid(row=0, column=2, padx=10, pady=10)

entry.bind("<KeyRelease>", update_word_count)

root.mainloop()
```
