```python
from IPython.display import display, HTML
def show_html(src):
    return display(HTML(src), metadata=dict(isolated=True))
```

```python
def mkdiv(content, **kwargs):
    if 'position' not in kwargs:
        kwargs['position'] = 'absolute'
    style_str = ' '.join([": ".join((k.replace('_', '-'), v))+";" for k, v in kwargs.items()])
    return f"<div style='{style_str}'>{content}</div>"
```

```python
import math
from IPython.display import display, HTML

divs = []
src = "The world is full of a number of things. I'm sure we should all be as happy as kings."

sentences = [sentence.strip() for sentence in src.split('.') if sentence.strip()]

radius = 150  # Adjusted for better visibility
center_x, center_y = 50, 50 

for i, sentence in enumerate(sentences):
    angle = i * (360 / len(sentences))
    radians = math.radians(angle)

    # Calculate text position relative to the center
    x_position = center_x + (radius * math.cos(radians))
    y_position = center_y + (radius * math.sin(radians)

    rotation_angle = angle

    this_div = mkdiv(
        sentence,
        position="absolute",
        top=f"{y_position}%",
        left=f"{x_position}%",
        transform=f"translate(-50%, -50%) rotate({rotation_angle}deg)",
        transform_origin="center",
        font_size="16px",
        text_align="center",
        white_space="nowrap"
    )
    divs.append(this_div)

html_src = html_tmpl.format(title="Circular Text", content="".join(divs))
display(HTML(html_src))
```

```python
import math

divs = []
n = 5  
src = ["The", "world", "is", "full", "of", "a", "number", "of", "things"]

radius_x = 30
radius_y = 30  
center_x, center_y = 50, 50 

for i in range(len(src)):
    word = src[i]
    
    angle = i * (360 / len(src))
    radians = math.radians(angle)

    x_position = center_x + radius_x * math.cos(radians)
    y_position = center_y + radius_y * math.sin(radians)

    rotation_angle = angle + 90

    this_div = mkdiv(
        word,
        position="absolute",
        top=f"{y_position}vh",
        left=f"{x_position}vh",

        height="auto",
        font_size="32px",
        text_align="center",
        transform=f"translate(-50%, -50%) rotate({rotation_angle}deg)"  # Rotate text to align with the circle
    )
    divs.append(this_div)

html_src = html_tmpl.format(title="circle", content="".join(divs))
show_html(html_src)
```
